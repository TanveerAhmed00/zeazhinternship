```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Specify columns to check for outliers
columns_to_check = [
    'Age of Subject', 'Time spent on Online Class', 'Time spent on self study', 
    'Time spent on sleep', 'Time spent on social media'
]

# Detecting Outliers Using IQR Method
def detect_outliers_iqr(df, columns):
    outliers = {}
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_indices = df[(df[column] < lower_bound) | (df[column] > upper_bound)].index
        outliers[column] = outlier_indices.tolist()
    return outliers

# Detect and display outliers
outliers_iqr = detect_outliers_iqr(df, columns_to_check)
print("Outliers detected using IQR method:")
print(outliers_iqr)

# Set up the Seaborn style and color palette
sns.set_theme(style="whitegrid")  # Choose a theme: "darkgrid", "whitegrid", "dark", "white", or "ticks"
palette = sns.color_palette("viridis", len(columns_to_check))

# Plotting boxplots with enhanced visuals
for i, column in enumerate(columns_to_check):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[column], color=palette[i], width=0.6, fliersize=5)
    plt.title(f'Distribution and Outliers for {column}', fontsize=14, weight='bold', color='darkblue')
    plt.xlabel(column, fontsize=12, weight='bold', color='gray')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Remove Outliers Using IQR Method
def remove_outliers_iqr(df, outliers):
    indices_to_drop = set()  # Collect unique indices across all columns
    for indices in outliers.values():
        indices_to_drop.update(indices)  # Use a set to avoid duplicates
    # Drop all unique indices
    df = df.drop(index=list(indices_to_drop), errors='ignore')
    return df

# Automatically remove outliers using IQR
df_cleaned = remove_outliers_iqr(df, outliers_iqr)
print("\nData after removing outliers using IQR method:")
print(df_cleaned)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# Assuming df_cleaned is already loaded and contains the necessary columns
# Standardize the 'Prefered social media platform' column to handle case variations
df_cleaned['Prefered social media platform'] = df_cleaned['Prefered social media platform'].str.strip().str.capitalize()

# Count plot with values on bars and a custom color palette from dark purple to light purple
plt.figure(figsize=(12, 8))

# Custom purple color palette from dark purple to light purple
custom_palette = sns.color_palette("Purples_d", n_colors=len(df_cleaned['Prefered social media platform'].unique()))

ax = sns.countplot(x='Prefered social media platform', data=df_cleaned, palette=custom_palette)

plt.title("Distribution of Preferred Social Media Platform", fontsize=18, weight='bold', pad=20)
plt.xlabel("Social Media Platform", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=45, fontsize=12)

# Add count values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='center', 
                fontsize=12, weight='bold', color='black', 
                xytext=(0, 10), textcoords='offset points')  # Position above bars

# Calculate the highest and lowest counts
platform_counts = df_cleaned['Prefered social media platform'].value_counts()
max_count = platform_counts.max()
max_count_platform = platform_counts.idxmax()
min_count = platform_counts.min()
min_count_platform = platform_counts.idxmin()

# Add annotations below the plot to avoid title collision
plt.figtext(0.5, -0.05, f'Highest: {max_count_platform} ({max_count}) | Lowest: {min_count_platform} ({min_count})', 
            ha='center', fontsize=14, color='darkred', weight='bold', wrap=True)

# Calculate the average time spent on social media for each platform
avg_time_spent = df_cleaned.groupby('Prefered social media platform')['Time spent on social media'].mean().reset_index()

# Annotate the average time spent inside the bars with red font color
for p, platform in zip(ax.patches, avg_time_spent['Prefered social media platform']):
    # Find the corresponding average time spent
    avg_time = avg_time_spent[avg_time_spent['Prefered social media platform'] == platform]['Time spent on social media'].values[0]
    height = p.get_height()
    
    # Annotate the average time spent inside the bar vertically with red font color
    ax.annotate(f'{avg_time:.2f}', 
                (p.get_x() + p.get_width() / 2., height / 2),  # Inside the bar
                ha='center', va='center', fontsize=12, weight='bold', color='red')

# Add a legend for the average time spent with custom purple color matching the bars
handles = [mlines.Line2D([0], [0], marker='o', color='red', markerfacecolor=custom_palette[0], markersize=10, label='Average Time Spent (hrs)')]
plt.legend(handles=handles, title="Legend", fontsize=12, loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Load and preprocess your data
# Assuming df_cleaned is your DataFrame with demographic or behavior-based data
non_numeric_cols = df_cleaned.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_cols)

# Step 3: Encode non-numeric columns (if needed) or drop them
# Option 1: Drop non-numeric columns if they’re not needed
df_numeric = df_cleaned.drop(non_numeric_cols, axis=1)
print("numeric columns:", df_numeric)

# Step 1: Standardize the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)

# Step 2: Determine the optimal number of clusters (for K-means)
# Using the Elbow Method
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal Clusters')
plt.show()

# Step 3: Apply K-means with the chosen number of clusters (e.g., 3)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(df_scaled)

# Step 4: Dimensionality Reduction for Visualization (using PCA)
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(df_pca, columns=['PCA1', 'PCA2'])
df_pca['Cluster'] = clusters

# Step 5: Plot the Clusters
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df_pca, palette='Set2')
plt.title('Cluster Visualization with PCA')
plt.show()


# Step 6: Assign the Cluster labels back to the original DataFrame
df_cleaned['Cluster'] = clusters
numeric_columns = df_cleaned.select_dtypes(include=['number']).columns
# Step 7: Profiling each cluster by checking the mean of the original features
cluster_profiling = df_cleaned[numeric_columns].groupby('Cluster').mean()
print("Cluster Profiling (mean of each feature per cluster):")
print(cluster_profiling)
```
